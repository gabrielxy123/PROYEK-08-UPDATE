```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf

from tensorflow import keras
from tensorflow.keras import layers
```

```python
data_train_path = r'C:\Users\lenov\Dataset Proyek 08 (2)\Dataset Proyek 08\train'
data_test_path = r'C:\Users\lenov\Dataset Proyek 08 (2)\Dataset Proyek 08\test'
data_val_path = r'C:\Users\lenov\Dataset Proyek 08 (2)\Dataset Proyek 08\validation'
```

```python
img_width = 180
img_height =180 
```

```python
data_train = tf.keras.utils.image_dataset_from_directory(
    data_train_path,
    shuffle=True,
    image_size=(img_width, img_height),
    batch_size=32,
    validation_split=False)
```

```python
data_cat = data_train.class_names
```

```python
data_cat
```

```python
data_val = tf.keras.utils.image_dataset_from_directory(data_val_path,
                                                       image_size=(img_height,img_width),
                                                       batch_size=32,
                                                        shuffle=False,
                                                       validation_split=False)
```

```python
data_test = tf.keras.utils.image_dataset_from_directory(
data_test_path,
    image_size=(img_height,img_width),
    shuffle=False,
    batch_size=32,
    validation_split=False
)
```

```python
plt.figure(figsize=(10,10))
for image, labels in data_train.take(1):
    for i in range(9):
        plt.subplot(3,3,i+1)
        plt.imshow(image[i].numpy().astype('uint8'))
        plt.title(data_cat[labels[i]])
        plt.axis('off')
```

```python
from tensorflow.keras.models import Sequential
```

```python
data_train
```

```python
from tensorflow.keras.regularizers import l2

model = Sequential([
    layers.Rescaling(1./255),
    
    # Conv2D + MaxPooling blok 1
    layers.Conv2D(16, 3, padding='same', activation='relu'),
    layers.MaxPooling2D(),
    
    # Conv2D + MaxPooling blok 2
    layers.Conv2D(32, 3, padding='same', activation='relu'),
    layers.MaxPooling2D(),
    
    # Conv2D + MaxPooling blok 3
    layers.Conv2D(64, 3, padding='same', activation='relu'),
    layers.MaxPooling2D(),
    
    # Flatten dan Dropout
    layers.Flatten(),
    layers.Dropout(0.4),  # Dropout dengan rasio lebih besar untuk regularisasi
    
    # Dense layer 1 dengan L2 regularisasi
    layers.Dense(128, activation='relu', kernel_regularizer=l2(0.01)),
    
    # Dense layer output (gunakan softmax untuk multi-class atau sigmoid untuk binary)
    layers.Dense(len(data_cat), activation='softmax')  # Sesuaikan dengan jumlah kategori
])
```

```python
model.compile(optimizer='adam', loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True), metrics=['accuracy'])
```

```python
from tensorflow.keras.callbacks import EarlyStopping,ReduceLROnPlateau

# Definisi callback EarlyStopping
early_stop = EarlyStopping(
    monitor='val_loss',  # Parameter yang dimonitor
    patience=5,          # Berhenti jika val_loss tidak membaik setelah 5 epoch
    restore_best_weights=True
)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, min_lr=1e-6)

history = model.fit(
    data_train,
    validation_data=data_val,
    epochs=20,
    callbacks=[early_stop, reduce_lr]
)
```

```python
# Ekstrak data dari history
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs_range = range(len(acc))

# Membuat grafik akurasi dan loss
plt.figure(figsize=(14, 6))

# Plot Akurasi
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Training Accuracy')
plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')

# Plot Loss
plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')

plt.show()
```

```python
image = 'contoh.jpg'
image = tf.keras.utils.load_img(image, target_size=(img_height,img_width))
img_arr = tf.keras.utils.array_to_img(image)
img_bat=tf.expand_dims(img_arr,0)
```

```python
predict = model.predict(img_bat)
```

```python
score = tf.nn.softmax(predict)
```

```python
print('Veg/Fruit in image is {} with accuracy of {:0.2f}'.format(data_cat[np.argmax(score)],np.max(score)*100))
```
